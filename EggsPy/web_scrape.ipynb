```python
#Importing dependencies
from splinter import Browser
from bs4 import BeautifulSoup as soup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
browser = Browser('chrome')
```

```python
# Visit the eggs
url = 'https://awionline.org/content/2019-barn-fire-statistics-state'
browser.visit(url)
```

```python
# Create a Beautiful Soup object
html = browser.html
html_fire19 = soup(html, 'html.parser')
```

```python
# Extract all the text elements
egglements = html_fire19.find_all('table', class_='table-barn-fires')
print(egglements)
```

```python
#List for headers
headers_list = []
for headers in egglements.find('th'):
    headers_list.append(headers.text)
print(headers_list)
```
